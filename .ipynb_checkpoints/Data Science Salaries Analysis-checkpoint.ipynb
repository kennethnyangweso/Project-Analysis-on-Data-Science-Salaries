## Feature Engineering
Feature engineering involves transforming raw data into meaningful features that enhance the predictive power and insights of our analysis.

```python
#Checking the format for the employee resident columns
ds1["Employee_Residence"]=ds1["Employee_Residence"]
print(ds1["Employee_Residence"].unique())

```

From the observation above it is evident that the column format are in country codes. This may be hard for analysis as stalkholders may not be familiar for country codes for each country around the world. To make this easy I will change the format to country name.

```python
import pycountry

# Function to check country code format and convert it
def country_code_to_name(code):
    if isinstance(code, str):  # Ensure it's a string
        code = code.strip().upper()  # Remove spaces and convert to uppercase
        
        # Check for 2-letter codes (ISO Alpha-2)
        country = pycountry.countries.get(alpha_2=code)
        if country:
            return country.name

        # Check for 3-letter codes (ISO Alpha-3)
        country = pycountry.countries.get(alpha_3=code)
        if country:
            return country.name
        
    return "Unknown"  # If no match is found

# Apply the function to convert country codes to names
ds1["Employee_Residence"] = ds1["Employee_Residence"].astype(str).apply(country_code_to_name)

# Display the first few rows
print(ds1["Employee_Residence"].value_counts())  # Check results


```

## Exploratory Data Analysis (EDA)
Here we will do visualization based on the business question and objectives

### What is the distribution in terms of popularity and salary accross different employment types:
Here I will perform an analysis how the employment types have been distributed in terms of their popularity and salary.

#### Univariate analysis:Employment popularity based on their count


```python
# Count the occurrences of each Experience Level and sort them
employment_type_order = ds1['Employment_Type'].value_counts().index

# Plotting the sorted count of each Experience Level
plt.figure(figsize=(8, 6))
sns.countplot(x='Employment_Type', data=ds1, order=employment_type_order)
plt.title('Count of Each Employment type')
plt.xlabel('Employment Type')
plt.ylabel('Count')
plt.show()
```

##### Observations

1. **Full-Time Employment Dominates** – The vast majority of individuals in the dataset are Full-Time employees, suggesting that this is the most common form of employment.

2. **Minimal Representation of Other Categories** – Part-Time, Contract, and Freelance employment types have extremely low counts, indicating they are rare in this dataset.

3. **Potential Data Collection Bias** – The dataset might be skewed towards Full-Time employees, either due to the nature of the population surveyed or how the data was collected.

4. **Limited Insights for Non-Full-Time Employees** – Since Part-Time, Contract, and Freelance categories have such low representation, drawing meaningful conclusions about them could be difficult.

#### Bi-variate analysis: Average salary Vs Employment type 

```python
# Grouping by Experience Level and calculating the average salary
avg_salary_by_type = ds1.groupby('Employment_Type')['Salary_In_Usd'].mean().sort_values(ascending=False)

# Plotting the bar plot for Experience Level vs Salary (sorted from largest to smallest)
plt.figure(figsize=(8, 6))
sns.barplot(x=avg_salary_by_type.index, y=avg_salary_by_type.values)
plt.title('Average Salary by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Average Salary')
plt.xticks(rotation=0)
plt.show()
```

##### Observations

1. **Full-Time Employees Have the Highest Average Salary** – They earn the most on average, significantly higher than other employment types.
2. **Contract Workers Earn Less Than Full-Time but More Than Others** – The average salary for contract workers is lower than full-time but still relatively high compared to part-time and freelance workers.
3. **Part-Time and Freelance Workers Earn the Least** – Part-time workers earn more than freelancers, but both have substantially lower average salaries than full-time and contract workers.

**Possible Explanation:** Hours and Benefits – The salary difference may be due to differences in working hours, job stability, and benefits. Full-time employees may receive higher pay due to job security and benefits, while freelance and part-time workers might earn less due to project-based or hourly work.

1. Full-time roles offer the highest salaries, especially for Data Scientists, followed by Data Engineers and Data Analysts.
2. Contract work remains competitive, particularly for Data Scientists and Data Analysts, though slightly lower than full-time.
3. Freelance Data Engineers earn significantly less, while Data Scientists still make a decent income in freelance roles.

### Are there trends in salary growth over the years?

#### Yearly trend for the top employment type 

```python
#univariate analysis based on popular employment type in terms of salary
# Filter for the employment type of interest
highest_type = "Full-Time"
fulltime = ds1[ds1['Employment_Type'].str.contains(highest_type, na=False)]

# Perform univariate analysis: Focus on average ratings over time
performance_year = fulltime.groupby('Work_Year')['Salary_In_Usd'].mean()

# Plotting the univariate trend
plt.figure(figsize=(8, 6))
performance_year.plot(kind='line', marker='o', color='blue')
plt.title("Salary performance for full time employees over the Years")
plt.xlabel("Year", fontsize=12)
plt.ylabel("Salary", fontsize=12)
plt.grid(visible=True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.xticks(ds1['Work_Year'].unique().astype(int))  # Set x-axis labels to integer years

plt.show()
```

##### Observations

1. **2020 to 2021:**

- There was a slight decline in salary from around $108,000 to $99,000.
- This could be due to external economic factors like the COVID-19 pandemic affecting salaries.

2. **2021 to 2022:**

- A significant salary increase to approximately $135,000 occurred.
- This suggests a strong recovery in the job market and possibly inflation-driven adjustments.

3. **2022 to 2023:**

- Another salary increase took place, reaching about $155,000.
- This indicates continued salary growth and demand for full-time employees.

4. **2023 to 2024:**

- There was a slight decline from $155,000  to around  $153,000.
- This suggests a stabilization or slight correction in salary growth, possibly due to market saturation or economic adjustments.

**Overall Insights:**

The general trend shows a strong recovery and growth from 2021 to 2023, with salaries peaking in 2023.
The minor decline in 2024 could indicate an economic adjustment rather than a major downturn.
Key influencing factors could include inflation, industry demand, and economic conditions.

2. Slight Decline (2024): The salary peaked in 2023, but there was a slight drop in 2024. This could be due to economic factors, industry shifts, or market saturation.

Overall, the trend suggests strong growth in compensation for data engineers, with a minor dip in the latest year.

## Conclusions

1. **Impact of Experience Level on Salary Variations**

- Salaries tend to increase with experience, as seen in the upward trend for specific roles.
- Senior positions generally earn significantly more than entry-level jobs.

2. **Salary Distribution Across Different Job Titles**

- Certain job titles, such as Data Engineers and Senior Data Scientists, consistently earn higher salaries.
- Specialized roles (e.g., Machine Learning Engineers) tend to have higher pay than generalist roles like Data Analysts.

3. **Effect of Employment Type on Salary**

- Full-time employees generally earn more than contract or freelance workers.
- Contract roles may have higher hourly rates, but full-time positions offer stability and benefits that contribute to long-term earnings.

4. **Influence of Company Size on Salary**

- Larger companies typically offer higher salaries due to larger budgets and competitive compensation strategies.
- Startups may provide lower base salaries but compensate with stock options or bonuses.

5. **Geographical Salary Disparities**

- Salaries vary significantly based on company location, with countries like Qatar, Israel, and the U.S. offering the highest compensation.
- Employee residence also plays a role, with remote work potentially bridging salary gaps between different regions.

6. **Salary Growth Over Time**

- The industry has seen a general upward trend in salaries, although some fluctuations exist.
- Economic conditions and industry demand influence salary trends, as seen in occasional drops.
